# Mission 1

What's the lenght of the following string ?

````
s = "Taumata whakatangihanga koauau o tamatea turi pukaka piki maungah oronuku pokai whenuaki tanatahu "

# affiche la longueur de la chaîne
print(len(s))
````

Now, count only non-space characters and display how many there is.
Remember, using a for loop on a string will browse all the individual characters of it!

````
s = "Taumata whakatangihanga koauau o tamatea turi pukaka piki maungah oronuku pokai whenuaki tanatahu "

# on initie un compteur à 0
compteur = 0

# la boucle va parcourir la chaîne de caractères
for i in s:
  # comme on veut tout récupérer sauf les espaces on créé une condition
  if i != " ":
    # si la condition est remplie on ajoute 1 au compteur
    compteur += 1

print(compteur)



````

# Mission 2

Ask the user for a string, then for a character (we assume user input follows the rules). Then, display the positions, **starting with 1**, of all occurrences of the character within the string.

For example, for the word "welcome" and for the character "e", the script will display:
```
position: 2
position: 7
```

````
string = input("Entrez une string : ")
char = input("Entrez un caractère : ")

# avec la fonction enumerate() la fonction va à chaque tour récupérer l'indice et la valeur de string
for index, valeur in enumerate(string):
  # on créé une condition pour isoler le caractère recherché
  if valeur == char:
    # on affiche l'indice (la position dans la chaîne)
    # comme on veut commencer à 1 et que les indices commencent à 0 on doit rajouter 1 à l'indice
    print(f"Position : {index + 1}")

````

# Mission 5

Ask the user for a string, then for a character, then for another character.

Then, display the string where the occurences of the first character are replaced by the second one.  
For example:
* string = `"Boulgour"`  
* 1st_char = `"g"`
* 2nd_char = `"v"`

Result : `"Boulvour"`

````
string = str(input())
char1 = str(input("1st char : "))
char2 = str(input("2nd char : "))

# la méthode replace() permet de remplacer un caractère ou une partie de chaîne
# doivent être renseigné l'élément à remplacer et l'élément qui le remplace
# on peut très bien remplacer par une chaîne vide ""
new_string = string.replace(char1, char2)

print(new_string)
````

# Mission 6

Ask the user for a string. Then, display the same string with all vowels converted to uppercase, and all consonants converted to lowercase.  
For example:
* string = `"Papa's got a brand new bag"`  
Result : `"pApA's gOt A brAnd nEw bAg"`


````
string = input("Entrez une phrase : ")

# plutôt que de faire un if très long je préfère créer une variable qui contient les voyelles
voyelles = "aeiouAEIOU"
# créé une chaîne vide qu'on va remplir au fur et à mesure
new_string = ""

# la variable i va parcourir les valeurs de string
for i in string:
  # condition pour récupérer les voyelles => si i est une voyelle, appliquer le traitement
  if i in voyelles:
    # la méthode join() récupére l'élément de string et l'ajoute à la chaîne vide
    # la méthode upper() met le caractère en majuscule
    new_string += "".join(i.upper())
  # sinon appliquer un autre traitement
  else:
    # la méthode lower() met le caractère en minuscule
    new_string += "".join(i.lower())

print(new_string)

````
