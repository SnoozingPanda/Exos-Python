```
# square of a number

def square_number(number):
  number = number**2
  return number

number = int(input())
square = square_number(number)
print(square)
```

```
# cube of a number

def cube_number(number):
  number = number**3
  return number

number = int(input())
cube = cube_number(number)
print(cube)
```

```
# absolute value of a number

def absolute_number(number):
  number = abs(number)
  return number

number = int(input())
absolute = absolute_number(number)
print(absolute)
```

```
# factorial of a number

def factorielle(number):
  if number == 0:
    return 1
  else:
    for i in range(1, number):
      number = number * i
  return number

number = int(input())
fact = factorielle(number)
print(fact)
```

```
# mode of a list of numbers

# def le_mode(liste_de_nombres):
from statistics import mode

def le_mode(liste_de_nombres):
  liste_de_nombres = max(set(liste_de_nombres), key = liste_de_nombres.count)
  print(f"Le mode de la liste est : {liste_de_nombres}")


liste_de_nombres =  [68, 99, 65, 44, 77, 44, 44]

combien = le_mode(liste_de_nombres)
print(combien)



```

```
# average of a list of numbers

def average(liste_de_nombres):
  somme = sum(liste_de_nombres)
  moyenne = round((somme / len(liste_de_nombres)), 2)
  print(f"La moyenne de la liste est de : {moyenne}")

liste_de_nombres = [68, 68, 68, 99, 65, 44, 77, 44, 44]
resultat = average(liste_de_nombres)
print(resultat)

```

```
# minimum of a list

liste_de_nombres = [68, 68, 68, 99, 65, 44, 77, 44, 44]

def le_plus_petit(liste_de_nombres):
  minimin = liste_de_nombres[0]
  for nombre in liste_de_nombres:
    if nombre < minimin:
      minimin = nombre
  return minimin

le_minimum = le_plus_petit(liste_de_nombres)
print(f"Le nombre minimum de la liste est : {le_minimum}")

```

```
# maximum of a list of numbers

liste_de_nombres = [68, 68, 68, 99, 65, 44, 77, 44, 44]

def le_plus_grand(liste_de_nombres):
  maximax = liste_de_nombres[0]
  for nombre in liste_de_nombres:
    if nombre > maximax:
      maximax = nombre
  return maximax

le_maximum = le_plus_grand(liste_de_nombres)
print(f"Le nombre maximum de la liste est : {le_maximum}")
```
