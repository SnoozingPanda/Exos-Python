# Mission 5

Create a function that returns the sum of two numbers when executed.  
Also, your function must return a sentence that simply indicates if the sum of the two numbers chosen by the user is:
* less than 10
* between 10 and 50
* greater than 50  
Again, you must use `return`.

```python
def calcul(a, b):
  somme = a + b
  if 10 < somme < 50:
    return "The sum of the two numbers chosen by the user is in between 10 and 50"
  elif somme < 10:
    return "The sum of the two numbers chosen by the user is less than 10"
  else:
    return "The sum of the two numbers chosen by the user is greater than 50"

a = int(input())
b = int(input())

calcul(a, b)

```
