# shorthand syntaxe

le principe de la shorthandsyntax est d'essayer d'ecrire un code en une seule ligne. ceci permet de rendre le code plus rapide en execution .

cependant faire attention à ne pas le faire quand votre code est complexe . c'est une pratique qui est intéressante dans la limite d'une code avec un for , if , else. si on a des elif , il n'est pas conseillé de faire la shorthand syntax

```python
#conditions : transformer en shorthand syntaxe (condition dans un print)
nombre = 7
if nombre % 2 == 0:
    print('chiffre pair')

else:
    print('chiffre impair')

```

```python
#en shorthand
nombre  = "chiffre pair" if nombre % 2 == 0 else 'chiffre impair'
print(nombre)
```

```python
#transformer en liste par comprehension
str_input = 'it is a good product !! 2023'
l_str = str_input.split()
list_str = []
for v in l_str:
    if v.isalpha():
        list_str += [v]
print(list_str)
```

```python
#list par comprehension
list_str = [ v for v in str_input.split() if v.isalpha()]
print(list_str)
```

```python
#list avec un for if else
#transformer en liste par comprehension
list_input = [5,7,9,10,12,15]
list_final = []
for v in list_input:
    if v % 2 == 0:
        list_final += [v // 2] #list_final.append(v//2)
    else:
        list_final += [v]

print(list_final)
```

```python
#list par comprehension
list_final = [ v// 2 if v % 2 == 0 else v  for v in list_input]
print(list_final)
```

```python
#transformer en dico par comprehension
dico = {'JANVIER' : 23, 'FEVRIER' : 25,'MARS': 18}
dico_test = {}
for k, v in dico.items():
    if v > 20 :
        dico_test[k] = v

print(dico_test)
```

```python
## shorthand syntaxe dictionnaire
dico_test = { k : v  for k, v in dico.items()  if v > 20}
print(dico_test)
```

```python
#transformer en dico par comprehension
#for if else
dico = {'JANVIER' : 23, 'FEVRIER' : 25,'MARS': 18}
dico_test = {}
for k, v in dico.items():
    if v > 20 :
        dico_test[k] = 'pb'
    else :
        dico_test[k] = v
print(dico_test)
```

```python
#ça devient complex pas une bonne idée ici de faire la short hand syntaxe
dico2 = {  k :  'pb' if v > 20 else v  for (k,v) in dico.items()}
print(dico2)
```

```python
#transformer en fct lambda
def div5(nombre):
    if nombre % 5 == 0:
        return nombre
    else :
        return nombre // 2
result = div5(7)
print(result)
```

```python
#fct lambda
div5 = lambda  nombre : nombre if nombre % 5 == 0 else nombre // 2
result = div5(7)
print(result)
```

```python
#transformer en shorthan syntax
def test(list_input = [5,7,9,10,12,15]):

    list_final = []
    for v in list_input:
        if v % 2 == 0:
            list_final += [v // 2] #list_final.append(v//2)
        else:
            list_final += [v]
    return list_final
r = test()
print(r)
```

```python
#shorthand syntax
test = lambda  list_input = [5,7,9,10,12,15]  : [v // 2 if v % 2 == 0 else v for v in list_input]
r = test()
print(r)
```

```python
#lambda pr modifier le comportement d'une fonction (parametre key):
list_string = ['azert', 'aze', 'azer',
               'azertttttttt', 'alekdjamdkf']

#ordonner par les chaine les plus longue vers les chaine moins longue
list_string = sorted(list_string,reverse = True, key= lambda x : len(x))
print(list_string)
```

```python
test = [("Lea" , 33), ("Loic",10), ("Henri",40)]

#ordonner par nom de personne
result = sorted(test,reverse = True, key= lambda x : len(x[0]))
print(result)
```
