# Mission 1

Define a list, then display only the elements (and their position) that have even positions (0, 2, 4, 6...).

With `test_list = ["a", "b", "c", "d", "e"]`, the script should display:
```
a at position 0
c at position 2
e at position 4
```

````
test_list = ["a", "b", "c", "d", "e"]

# enumerate retourne 2 choses : l'indice et l'élément de la liste
# on créé donc deux variables qui vont contenir l'indice et l'élément
for index, elem in enumerate(test_list):

  # on cherche les indices paires avec une condition
  if index % 2 == 0:
    print(f"{elem} at position {index}")
````

# Mission 2

Define a list, then **ask** for a position, then for a string. Then replace the list element at that position by the given string, and display the resulting list.

For example, with the list `test_list2 = ['hello', 'good morning', 'bye bye', 'have a good day']`, the position `2` and the string `yeah`, you should display:
```
['hello', 'good morning', 'yeah', 'have a good day']
```

````
test_list2 = ['hello', 'good morning', 'bye bye', 'have a good day']

position = int(input('position ? :'))

# la méthode pop() enlève l'élément à la position indiquée
test_list2.pop(position)
print(test_list2)

# la méthode insert() insert le nouvel élément à la position indiquée
# attention à la syntaxe: on indique la position puis l'élément à ajouter
test_list2.insert(2, 'yeah')
print(test_list2)







````

# Mission 3

Define a list with five numbers, then ask the user for a number. Put the number at the end of the list, remove the first element from the list, and finally display the list.

For example, with the list `test_list3 = [1, 9, 6, 15, 4]` and the input `7`, you should display:
```
[9, 6, 15, 4, 7]
```


````
test_list3 = [1, 9, 6, 15, 4]

user = int(input())

# ajoute l'élément en fin de liste
test_list3.append(user)
print(test_list3)

# enleve l'élément indiqué entre parenthèses
test_list3.remove(1)
print(test_list3)
````

# Mission 4

Define a list, then ask the user for a string. Then, insert the string in the list at a random position, and display the resulting list.

For example, with the list `test_list4 = ["p", "y", "t", "h", "o"]` and the input `n`, you *might* get:
```
["p", "n", "y", "t", "h", "o"]
```
... but you might also get:
```
["p", "y", "t", "n", "h", "o"]
```


````
import random

test_list4 = ["p", "y", "t", "h", "o"]

user = str(input())

# détermine une position aléatoire
position = random.randint(0,5)

# insert la chaîne créée par user à la position aléatoire créée par random.randint()
test_list4.insert(position, user)

print(test_list4)
````

# Mission 6

Ask the user for a string, convert it to a list of characters with the `list` function, then ask for a single character.

Remove the first occurrence of the character from the list, then display how many occurrences of the character remain in the list.

For instance if you enter `"Excellent work"` then the letter `"e"`, it will display:
```
There are still 2 copies of e in the list
```

````
# on demande à l'utilisateur d'entrée une chaîne
# on s'assure d'éviter les problèmes de casse en mettant tout en minuscule
user = input().lower()

# créé une liste à partir de l'input() et la stocke dans une variable
sequence = list(user)

# on demande à l'utilisateur d'entrer un caractère
caractere = input()

# on créé une liste vide
compteur = []

for elem in sequence:             # pour chaque élément de la séquence
  if elem == caractere:           # si l'élément est égal au caractère entré par l'utilisateur
    compteur.append(elem)         # on ajoute l'élément au compteur
    # print(compteur)

compteur.remove(caractere)        # enlève le 1er élément trouvé qui a la valeur demandée
copies = len(compteur)            # donne la longueur de la liste après le remove()

print(f"Il reste {copies} copie de '{caractere}' dans la liste")




````

# Mission 7

Define three lists containing no duplicates. Then, indicate (however you like) the two that have the most elements in common (or all three if there's a tie).

For instance, with:
```
list_a = [1, 2, 4, 8, 16, 32]
list_b = [1, 2, 3, 5, 8, 13]
list_c = [2, 3, 5, 7, 11, 13]
```
It may display:
```
Lists B and C have the most common elements!
```

````
list_a = [1, 2, 3, 4, 5, 6, 8, 11, 21]
list_c = [1, 2, 3, 7, 10, 19, 21, 23, 30]
list_b = [10, 11, 12, 13, 14, 15, 16, 17, 18]

# on créé des compteurs pour chaque paire de listes possibles
compteur_ab = 0
compteur_bc = 0
compteur_ac = 0

# une boucle qui va comparer la liste a avec la liste b
for elem in list_a:
  if elem in list_b:
    compteur_ab += 1
print(f"Le nombre d'éléments communs entre list_a et list_b : {compteur_ab}")
print("-----")

# une boucle qui va comparer la liste b avec la liste c
for elem in list_b:
  if elem in list_c:
    compteur_bc += 1
print(f"Le nombre d'éléments communs entre list_b et list_c : {compteur_bc}")
print("-----")

# une boucle qui va comparer la liste c avec la liste a
for elem in list_c:
  if elem in list_a:
    compteur_ac += 1
print(f"Le nombre d'éléments communs entre list_c et list_a : {compteur_ac}")
print("-----")

# on compare les compteurs pour savoir lequel
if compteur_ab > compteur_bc and compteur_ab > compteur_ac:
  print("Les listes A et B ont le plus d'éléments en commun")
elif compteur_bc > compteur_ab and compteur_bc > compteur_ac:
  print("Les listes B et C ont le plus d'éléments en communs")
else:
  print("Les listes A et C ont le plus d'éléments en commun")










````
